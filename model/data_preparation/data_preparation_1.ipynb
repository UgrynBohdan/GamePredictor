# Об'єднання потрібних даних

```python
games = pd.read_csv('../raw/games.csv')
clubs = pd.read_csv('../raw/clubs.csv')
competitions = pd.read_csv('../raw/competitions.csv')
```

```python
games_clubs_home = merge_tables(
    games,
    clubs,
    left_on='home_club_id',
    right_on='club_id',
    right_str='_home_club'
)
```

```python
games_clubs = merge_tables(
    games_clubs_home,
    clubs,
    left_on='away_club_id',
    right_on='club_id',
    right_str='_away_club'
)
```

```python
games_clubs_competitions = merge_tables(
    games_clubs,
    competitions,
    left_on='competition_id',
    right_on='competition_id'
)
```

```python
show_table_info(games_clubs_competitions, str_table='games_clubs_competitions', all_info=True)
```

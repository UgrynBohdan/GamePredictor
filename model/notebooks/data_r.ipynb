```python
df = pd.DataFrame()
```

```python
df[X] = None
```

# Заповнення df

### competition_id

```python
competition[competition['competition_code'] == example['competition_code']]['competition_id'].values[0]
```

```python
df['competition_id'] = competition[competition['competition_code'] == example['competition_code']]['competition_id'].values[0]
```

```python
df
```

### season, date

```python
df['season'] = pd.to_datetime(example['date']).year
```

```python
df['date'] = example['date']
```

```python
df
```

### round

```python
# найпоширеніший код
df['round'] = game['round'].mode()[0]
df
```

### home_club_id, away_club_id

```python
clubs[clubs['name'] == example['home_club_name']]['club_id'].values
```

```python
df['home_club_id'] = clubs[clubs['name'] == example['home_club_name']]['club_id'].values
df['away_club_id'] = clubs[clubs['name'] == example['away_club_name']]['club_id'].values
```

```python
df
```

### home_club_position, away_club_position

```python
df['home_club_position'] = game['home_club_position'].mode()[0]
df['away_club_position'] = game['away_club_position'].mode()[0]
df
```

### home_club_manager_name, away_club_manager_name

```python
df['home_club_manager_name'] = game['home_club_manager_name'].mode()[0]
df['away_club_manager_name'] = game['away_club_manager_name'].mode()[0]
df
```

### stadium

```python
df['stadium'] = example['stadium']
df
```

### attendance

```python
df['attendance'] = example['attendance']
df
```

### referee

```python
df['referee'] = example['referee']
df
```

### domestic_competition_id_home_club,  
### squad_size_home_club,  
### average_age_home_club,  
### foreigners_number_home_club,  
### national_team_players_home_club,  
### stadium_name_home_club,  
### stadium_seats_home_club,  
### last_season_home_club

```python
df[[
    'domestic_competition_id_home_club', 'squad_size_home_club',
       'average_age_home_club', 'foreigners_number_home_club',
       'national_team_players_home_club', 'stadium_name_home_club',
       'stadium_seats_home_club', 'last_season_home_club', 'net_transfer_record_home_club'
]] = clubs[clubs['club_id'] == df['home_club_id'].values[0]][[
    'domestic_competition_id', 'squad_size',
       'average_age', 'foreigners_number',
       'national_team_players', 'stadium_name',
       'stadium_seats', 'last_season', 'net_transfer_record'
]].values
```
